```python
import pandas as pd
import numpy as np
```

```python
for i in range(2022,2002,-1):
    print(f'<h5><a href = "#{i}">{i}</a></h5>')
```

```python
j_df = pd.read_csv('jennypubs.csv')
```

```python
v_df = pd.read_csv('vivekpubs.csv')
```

```python
F = [j_df,v_df]
```

```python
df = pd.concat(F)
df = df.drop_duplicates(subset = ['Authors','Title','Year']).sort_values(by = 'Year', ascending = False).reset_index(drop = True)
df = df.rename(columns = {'Source title':'Journal'})
df = df.loc[df['Document Type'].isin(['Article', 'Review', 'Editorial', 'Book Chapter', 'Conference Paper'])]

df['Author Keywords'].fillna(df['Index Keywords'], inplace=True)
df = df.reset_index(drop = True)
df['num'] = np.arange(len(df),0,-1)

pp = df.loc[df['Publication Stage'] == 'Article in Press']
df = df.loc[df['Publication Stage'] == 'Final']

```

```python
for year in df.Year.unique():
    print(df.loc[df.Year == year])
```

```python
def format_csv(df, pp = False):
    df = df[['num','Authors', 'Title', 'Year', 'DOI',  'Author Keywords', 'Document Type', 'Journal','Abstract', 'Publication Stage']]
    pub_string = ''
    for year in df.Year.unique():
        year_df = df.loc[df.Year == year]
        if pp == False :
            pub_string += f"<h3>{year}</h3>"
        else:
            pub_string += f"<h3>Preprint</h3>"
        pub_string += '<hr>'
        for _, pub in year_df.iterrows():
            if pp  == False:
                pubyear = f', {pub.Year}'
                pubnum = f'{pub.num}. '
            else:
                pubyear = ''
                pubnum = ''

            pub_string += f'''
            <h4><a href = "https://doi.org/{pub.DOI}" target = "_blank">{pubnum}{pub.Title}</a></h4>
            <div id = "authors"><p>{pub.Authors}</p></div>
            <div id = "journal"><p><i>{pub.Journal}{pubyear}</i></p></div>

            '''
    return pub_string
```

```python
publications_html = f'''{{{{ define "container" }}}}


<!-- <div class="container" > -->
  <!-- sidebar -->
  <div class="col-lg-3 col-lg-offset-0 col-md-3 col-md-offset-0 col-sm-12 col-xs-12 sidebar-container">
        <!-- Featured Tags -->
        {{{{ partial "pub_sidebar.html" . }}}}
  </div>
    
  <!-- publications -->
  <div class = "col-lg-8 col-lg-offset-1 col-md-8 col-md-offset-1 col-sm-12 col-xs-12 pubs-container" style = 'padding : 0; margin : 0;'>
  {format_csv(pp, pp  = True)}
  
  {format_csv(df)}
  </div>
</div>


  
{{{{end}}}}
'''
```

```python
f = open('../themes/puppet/layouts/section/publications.html', 'w')
f.write(publications_html)
f.close()
```
